```python
import pandas as pd
df = pd.read_csv('output.csv')
df.head()
```

```python
df = df.drop(["tweet_id", "lang", "verified", "sentiment_neutral", "entail_prob", "contra_prob"], axis = 1, inplace = True)
```

```python
df['stance'].unique()
```

```python
df.groupby("country")["stance"]

countries_dict = {}
probabilities_dict = {}
for index, row in df.iterrows():
    if row["country"] not in countries_dict:
        countries_dict[row["country"]] = [0, 0, 0] #[proUkr, proRus, unsure]
        probabilities_dict[row["country"]] = [0, 0, 0] #[proUkr, proRus, unsure]
    if row["stance"] == "pro-Ukraine":
        countries_dict[row["country"]][0] += 1
    elif row["stance"] == "pro-Russia":
        countries_dict[row["country"]][1] += 1
    elif row["stance"] == "unsure":
        countries_dict[row["country"]][2] += 1

countries_dict
```

```python
for key, value in countries_dict.items():
    country_labels = countries_dict[key]
    for i in range(3):
        probabilities_dict[key][i] = country_labels[i] / sum(country_labels)
probabilities_dict
```

```python
prob_df = pd.DataFrame(probabilities_dict).T
prob_df = prob_df.rename(columns={
    0: "pro-Ukraine",
    1: "pro-Russia",
    2: "unsure"
})
prob_df

prob_df.to_csv("probabilities.csv")
```
